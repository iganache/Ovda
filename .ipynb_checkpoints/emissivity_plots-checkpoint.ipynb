```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import EngFormatter, StrMethodFormatter
from mpl_toolkits.mplot3d import Axes3D
import math
import pandas as pd
import geopandas as gpd
import fiona
```

### <b> EMISSIVITY VS INCIDENCE ANGLE FOR VPOL AND HPOL DATA </b>

```python
ovda_v = gpd.read_file("/mnt/c/Users/iganesh/Work/Research/Venus/ImpactDeposits/Datasets/Ovda/Final/polarization/ovda_vpol_30km_buffer.gpkg", layer = "rdf")
ovda_h = gpd.read_file("/mnt/c/Users/iganesh/Work/Research/Venus/ImpactDeposits/Datasets/Ovda/Final/polarization/ovda_hpol_30km_buffer.gpkg", layer = "rdf")

dfh = pd.DataFrame(data = ovda_h).sort_values('RAD_FOOTPRINT_LATITUDE')
dfv = pd.DataFrame(data = ovda_v).sort_values('RAD_FOOTPRINT_LATITUDE')
# dfh["MEAN_SURFACE_EMISSIVITY"] = dfh["SURFACE_EMISSIVITY"].rolling(window=10).mean()
# dfv["MEAN_SURFACE_EMISSIVITY"] = dfv["SURFACE_EMISSIVITY"].rolling(window=10).mean()
```

```python
min_thi = dfh['INCIDENCE_ANGLE'].min()
max_thi = dfh['INCIDENCE_ANGLE'].max()

bins = np.arange(min_thi, max_thi+0.05, 0.1)
dfv = dfv[(dfv['INCIDENCE_ANGLE'] >= min_thi) & (dfv['INCIDENCE_ANGLE'] <= max_thi)]
```

```python
dfh["BIN"] = pd.cut(dfh["INCIDENCE_ANGLE"], bins, right=True)
dfh["MEAN_SURFACE_EMISSIVITY"] = dfh.groupby("BIN")["SURFACE_EMISSIVITY"].transform("mean") 
dfh["STD_SURFACE_EMISSIVITY"] = dfh.groupby("BIN")["SURFACE_EMISSIVITY"].transform("std") 
dfh["MEAN_INCIDENCE_ANGLE"] = dfh.groupby("BIN")["INCIDENCE_ANGLE"].transform("mean") 
dfh["MEAN_RAD_FOOTPRINT_LATITUDE"] = dfh.groupby("BIN")["RAD_FOOTPRINT_LATITUDE"].transform("mean") 

dfv["BIN"] = pd.cut(dfv["INCIDENCE_ANGLE"], bins, right=True)
dfv["MEAN_SURFACE_EMISSIVITY"] = dfv.groupby("BIN")["SURFACE_EMISSIVITY"].transform("mean") 
dfv["STD_SURFACE_EMISSIVITY"] = dfv.groupby("BIN")["SURFACE_EMISSIVITY"].transform("std") 
dfv["MEAN_INCIDENCE_ANGLE"] = dfv.groupby("BIN")["INCIDENCE_ANGLE"].transform("mean") 
dfv["MEAN_RAD_FOOTPRINT_LATITUDE"] = dfv.groupby("BIN")["RAD_FOOTPRINT_LATITUDE"].transform("mean") 
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(8, 4)

# plt.grid(visible=True, which='major', axis='both', linestyle="--", linewidth=0.5)

matplotlib.rcParams['font.sans-serif'] = "Arial"
matplotlib.rcParams['font.family'] = "sans-serif"

hcolor = "#ae76a3"
vcolor = "#90c987"

hcoloravg = "#882e72"
vcoloravg = "#4eb27f"

xfield = 'RAD_FOOTPRINT_LATITUDE'
# xfield = 'INCIDENCE_ANGLE'

ax.axvline(-4.85, color = "#560133", linewidth = 0.8, linestyle="--" )
ax.axvline(-7.77, color = "#560133", linewidth = 0.8, linestyle="--")
ax.axvline(-4.77, color = 'k', linewidth = 0.8, linestyle="--")
ax.axvline(3.48, color = 'k', linewidth = 0.8, linestyle="--")

dfh.plot.scatter(xfield, 'SURFACE_EMISSIVITY', ax=ax, s = 10, color = "white", edgecolor = hcolor)
dfv.plot.scatter(xfield, 'SURFACE_EMISSIVITY', ax=ax, s = 10, color = "white", edgecolor = vcolor)

ax.fill_between(dfh['MEAN_RAD_FOOTPRINT_LATITUDE'], dfh['MEAN_SURFACE_EMISSIVITY'] - dfh['STD_SURFACE_EMISSIVITY'], dfh['MEAN_SURFACE_EMISSIVITY'] + dfh['STD_SURFACE_EMISSIVITY'], color = hcolor, alpha = 0.5)
ax.fill_between(dfv['MEAN_RAD_FOOTPRINT_LATITUDE'], dfv['MEAN_SURFACE_EMISSIVITY'] - dfv['STD_SURFACE_EMISSIVITY'], dfv['MEAN_SURFACE_EMISSIVITY'] + dfv['STD_SURFACE_EMISSIVITY'], color = vcolor, alpha = 0.5)

dfv.plot.scatter('MEAN_RAD_FOOTPRINT_LATITUDE', 'MEAN_SURFACE_EMISSIVITY', ax=ax, s = 15, label="V-polarized",  marker = "x", color = vcoloravg)
dfh.plot.scatter('MEAN_RAD_FOOTPRINT_LATITUDE', 'MEAN_SURFACE_EMISSIVITY', ax=ax, s = 15, label="H-polarized",  marker = "x", color = hcoloravg)



ax.set_xlabel("Footprint center latitude", fontsize = 10)
ax.set_ylabel("Polarized emissivity", fontsize = 10)
ax.set_xlim(-10, 7)
ax.set_ylim(0.25, 1.05)
ax.tick_params(axis="both", which="major", labelsize=10)
ax.xaxis.set_major_formatter(StrMethodFormatter(u"{x:.0f} "))
ax.xaxis.labelpad = 8
ax.yaxis.labelpad = 8
ax.legend(fontsize = 10, loc = 2)

ax.set_xlim(-10, 4.5)
plt.grid(visible=True, which='major', axis='both', linewidth=0.5)

# plt.savefig("polarized_emissivity_grid.pdf", bbox_inches='tight')
plt.show()
```

### <B> ELEVATION NOISE REMOVAL FUNCTION AND USAGE </B>

```python
ovda_old = gpd.read_file("/mnt/c/Users/iganesh/work/Research/Venus/ImpactDeposits/Datasets/Ovda/Final/gpkg/Mar26/arcdr_ovda_cycle1_Mar26_no_ITP.gpkg", layer = "rdf")
ovda_new_filename = "/mnt/c/Users/iganesh/work/Research/Venus/ImpactDeposits/Datasets/Ovda/Final/gpkg/Mar26/arcdr_ovda_cycle1_Mar26_no_ITP_05deg_n50_lowsig2_highsig2_1degoverlap_sigemis2.gpkg"
ovda_new = ovda_old.copy()
```

```python
def remove_noise(ogdf, newdf):
    minx, miny, maxx, maxy = ogdf.total_bounds
    x = np.arange(minx, maxx, 0.5)
    y = np.arange(miny, maxy, 0.5)

    for i in x:
        for j in y:
            try:
                roi = ogdf.cx[i:i+1, j:j+1]
            except:
                roi = ogdf.cx[i:maxx, j:maxy]
            if len(roi) >= 50:
                plus_3sigma = roi["AVERAGE_PLANETARY_RADIUS"].mean() + 2*roi["AVERAGE_PLANETARY_RADIUS"].std()
                minus_3sigma = roi["AVERAGE_PLANETARY_RADIUS"].mean() - 2*roi["AVERAGE_PLANETARY_RADIUS"].std()
                plus_3sigmaemis = roi["SURFACE_EMISSIVITY"].mean() + 2*roi["SURFACE_EMISSIVITY"].std()
                minus_3sigmaemis = roi["SURFACE_EMISSIVITY"].mean() - 2*roi["SURFACE_EMISSIVITY"].std()
                # elev_noise = roi[(roi["AVERAGE_PLANETARY_RADIUS"] < minus_3sigma) | (roi["AVERAGE_PLANETARY_RADIUS"] > plus_3sigma)]
                elev_noise = roi[((roi["AVERAGE_PLANETARY_RADIUS"] < minus_3sigma) | (roi["AVERAGE_PLANETARY_RADIUS"] > plus_3sigma)) | ((roi["SURFACE_EMISSIVITY"] < minus_3sigmaemis) | (roi["SURFACE_EMISSIVITY"] > plus_3sigmaemis))]
                try:
                    newdf.drop(elev_noise.index.to_list(), inplace = True)
                except:
                    pass
```
